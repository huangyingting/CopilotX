## Underwriting in Health Insurance
The underwriting process in health insurance is a critical function that involves evaluating an applicant's risk to determine eligibility and premium rates. This often requires sifting through extensive medical histories, interpreting potentially conflicting information, and applying dynamic underwriting guidelines. 

This example Demonstrates how a reasoning model could assist in the underwriting process.

```python
DEVELOPER_PROMPT = """
Use clear headings, bullet points for lists, and tables for any numerical data where applicable. Respond in markdown and don't use LaTeX format.
"""

USER_PROMPT = """
<background>

### `Doc_Application_JD001`: Applicant's Signed Application Form
- **Applicant ID:** JD-2025-05-26-001
- **Name**: John Doe
- **Age**: 45
- **Gender**: Male
- **Policy Applied For**: Term Life Insurance
- **Coverage Amount**: $500,000
- **Occupation**: Office Manager
- **Smoker Status (Self-Declared on Application)**: "No, never smoked."
- **Disclosed Medical Conditions (on Application)**:
    - Hypertension (diagnosed age 40)
    - High Cholesterol (diagnosed age 42)
- **Family History (Disclosed on Application)**:
    - Father: Heart attack at age 55
    - Mother: Type 2 Diabetes
- **Policy Applied For**: Term Life Insurance, $500,000
- Respiratory Health Questionnaire: Answered "No" to questions about asthma, bronchitis, inhaler use, or other lung conditions.

### `Doc_APS_Summary_JD001`: Attending Physician Statement Summary - Dr. Emily Carter, dated 3 months ago
- Hypertension: Diagnosed 5 years ago. Current medication: Lisinopril 10mg daily. Last BP reading: 135/85 mmHg. Noted as "stable and managed."
- Hyperlipidemia (High Cholesterol): Diagnosed 3 years ago. Current medication: Atorvastatin 20mg daily. Last LDL cholesterol: 110 mg/dL. Noted as "well-controlled."
- Family History: Confirms father's MI at 55, mother's T2DM.
- Past Notes: Brief mention of "patient reported occasional anxiety symptoms" 7 years ago during a routine check-up; no specific diagnosis made, no treatment prescribed, no specialist referral, no follow-up notes on this issue.
- Lifestyle: Patient reports regular exercise (3x/week) and a balanced diet.
- Respiratory: No active respiratory complaints noted in recent visits. Lungs clear on auscultation at last physical.
- Tobacco Use (as per medical record): Note from 6 years ago: "Patient reports smoking 5-10 cigarettes/day, advised cessation." Note from 4 years ago: "Patient states successfully quit smoking 2 years ago." (This conflicts with "never smoked" on application).

### `Doc_Pharmacy_Report_JD001`: Pharmacy Benefit Manager Report - Last 12 months
- Lisinopril 10mg: Regular refills.
- Atorvastatin 20mg: Regular refills.
- Albuterol Inhaler (90mcg): Filled twice in the last 12 months (most recent fill 2 months ago). (Not mentioned on application or in APS summary).

### `Doc_UW_Guide_Hypertension_v2.3`: Insurer's Underwriting Guideline - Hypertension
- Well-controlled (BP < 140/90 mmHg with meds): Standard to +10% rating, depending on age of onset and other risk factors.
- Moderately controlled (BP 140-159/90-99 mmHg): +25% to +50% rating.
- Poorly controlled (BP >= 160/100 mmHg): Decline or significant rating (+75%+) with senior underwriter review.

### `Doc_UW_Guide_Cholesterol_v1.8`: Insurer's Underwriting Guideline - Cholesterol
- Well-controlled (LDL < 130 mg/dL with meds): Standard rates.
- Elevated (LDL 130-159 mg/dL): +15% rating.
- High (LDL >= 160 mg/dL): +25% to +50% rating.

### `Doc_UW_Guide_Respiratory_v3.1`: Insurer's Underwriting Guideline - Respiratory Conditions:
- Undisclosed inhaler use (e.g., Albuterol): Requires investigation. If for occasional, infrequent viral bronchitis (max 1-2 times/year, no underlying asthma), may be standard. If indicative of undisclosed asthma or chronic bronchitis:
    - Mild intermittent asthma (well-controlled, infrequent inhaler use): +25% to +50% rating.
    - Moderate/Severe asthma: Higher ratings or decline.
- Must clarify reason for use.

### `Doc_UW_Guide_FamilyHistory_v1.2`: Insurer's Underwriting Guideline - Family History
- Cardiovascular Event in First-Degree Male Relative < 60 yrs: +15% rating.
- Type 2 Diabetes in First-Degree Relative: +5% rating if applicant has other metabolic risk factors (e.g., BMI > 28, pre-diabetes – not present here based on data). Otherwise, note for context.

### `Doc_UW_Guide_TobaccoUse_v4.0`: Insurer's Underwriting Guideline - Tobacco Use
- Current Smoker or Quit < 1 year ago: Smoker rates (typically +100% to +200% over non-smoker).
- Quit 1-3 years ago: Preferred Smoker rates (typically +50% to +75%).
- Quit 3-5 years ago: Standard Plus Non-Smoker rates (typically +25%).
- Quit > 5 years ago: Standard Non-Smoker rates (if no other major risks).
- Misrepresentation of smoking history (if discovered): Re-underwrite at appropriate smoker classification based on actual quit date. If within contestability period, consider policy voiding.

### `Doc_UW_Guide_Anxiety_v2.5`: Insurer's Underwriting Guideline - Mental Health
- Isolated mention of anxiety symptoms, no diagnosis, no treatment, >5 years ago: Typically no rating. Note for file.
- Active diagnosis or treatment: Refer to specific guidelines based on severity.

### `Doc_UW_Rating_Calc_Rules_v1.0`: Insurer's Underwriting Rating Calculation Rules
- Ratings are generally additive unless total exceeds +100%, then refer to senior underwriter.
- Maximum allowable rating before decline (for this product): +150%.
- Misrepresentation findings supersede initial calculations and require re-evaluation from base.
</background>

<instructions>
1.  **Information Synopsis & Discrepancy Identification:**
    a.  Summarize key information from the application (`Doc_Application_JD001`).
    b.  Summarize key medical findings from the APS (`Doc_APS_Summary_JD001`).
    c.  Note any additional relevant information from the Pharmacy Report (`Doc_Pharmacy_Report_JD001`).
    d.  Identify and clearly state any discrepancies or missing information between these sources (especially between self-declared information and information from other documents). Focus on tobacco use and respiratory health.

2.  **Risk Factor Assessment (Apply Guidelines):**
    For each potential risk factor identified, perform the following:
    a.  **Hypertension:** Assess control status and determine rating based on `Doc_UW_Guide_Hypertension_v2.3`.
    b.  **High Cholesterol:** Assess control status and determine rating based on `Doc_UW_Guide_Cholesterol_v1.8`.
    c.  **Family History:** Determine rating based on `Doc_UW_Guide_FamilyHistory_v1.2`.
    d.  **Past Anxiety Mention:** Assess significance based on `Doc_UW_Guide_Anxiety_v2.5`.
    e.  **Tobacco Use History:** Based on the discrepancy noted, determine the correct smoker classification and associated rating impact according to `Doc_UW_Guide_TobaccoUse_v4.0`. State the implication of the misrepresentation.
    f.  **Undisclosed Albuterol Use:** Assess potential risk and determine necessary actions/preliminary rating impact based on `Doc_UW_Guide_Respiratory_v3.1`.

3.  **Preliminary Rating Calculation:**
    a.  Starting from a base of Standard Non-Smoker (unless reclassified due to tobacco misrepresentation).
    b.  Sum the percentage ratings determined for each risk factor above, considering the `Doc_UW_Rating_Calc_Rules_v1.0`.
    c.  State the total preliminary percentage rating.

4.  **Overall Assessment and Next Steps:**
    a.  Provide a brief overall assessment of the applicant's risk profile.
    b.  Recommend specific next steps for the human underwriter (e.g., request specific clarifications from the applicant, order additional tests, refer to senior underwriter). Justify each recommendation.

**Output Format:**
A structured report with clear headings for each of the above sections (1, 2, 3, 4) and sub-sections (a, b, c, etc.). All conclusions and ratings must be explicitly justified by referencing the specific guideline document and relevant clause/section.    
</instructions>

<question>
You are an AI Underwriting Risk Assessment Specialist. Your task is to analyze the provided applicant data and retrieved documents to perform a preliminary risk assessment for a Term Life Insurance application. You must identify all relevant risk factors, flag any discrepancies or missing information, apply the insurer's underwriting guidelines, calculate a preliminary rating, and recommend next steps. 
</question>
"""
```

```python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

load_dotenv()

# Options: low, medium, high
REASONING_LEVEL = "high"
# auto, concise, or detailed (currently only supported with o4-mini and o3)
SUMMARY_LEVEL = "detailed"

# Set up the Azure OpenAI client
client = AzureOpenAI(
    azure_endpoint=os.getenv("AZURE_OPENAI_ENDPOINT"),
    api_key=os.getenv("AZURE_OPENAI_KEY"),
    api_version="2025-04-01-preview"
)

response = client.responses.create(
    model="o4-mini",
    input=[
        {"role": "developer", "content": DEVELOPER_PROMPT},
        {"role": "user", "content": USER_PROMPT},
    ],
    reasoning={
        "effort": REASONING_LEVEL,
        "summary": SUMMARY_LEVEL
    }
)
```

### Result

```python
from IPython.display import display, Markdown

for item in response.output:
  if getattr(item, "type", None) == "message" and hasattr(item, "content"):
    display(Markdown(item.content[0].text))
```

### Reasoning Summary

```python
from IPython.display import display, Markdown

for item in response.output:
  if getattr(item, "type", None) == "reasoning" and hasattr(item, "summary"):
    for summary in item.summary:
      if getattr(summary, "type", None) == "summary_text":
        display(Markdown(summary.text))
```
